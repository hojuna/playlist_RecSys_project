```python
import os
from argparse import ArgumentParser
from typing import List, Tuple

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from mlp_baseline.datasets import MLPDataset
from mlp_baseline.models import MLPModel
from scipy.io import mmread
from torch.utils.data import DataLoader
from tqdm import tqdm

train_matrix = mmread("data/split_data/train_matrix.mtx").tocsr()
valid_matrix = mmread("data/split_data/valid_matrix.mtx").tocsr()

# 데이터셋 생성
train_dataset = MLPDataset(train_matrix, num_negatives=4)
valid_dataset = MLPDataset(valid_matrix, num_negatives=4)
```

```python
import csv
song_names = {}
with open("data/song_info_50_50.csv") as f:
    reader = csv.reader(f)
    next(reader)
    for song_id, song_name, artist_name in reader:
        song_names[int(song_id)] = (song_name, artist_name)
```

```python
user_id, pos_song_ids, neg_song_ids = train_dataset[10]

for pos_song_id in pos_song_ids:
    pos_song_id = int(pos_song_id)
    print(song_names[pos_song_id])
```
